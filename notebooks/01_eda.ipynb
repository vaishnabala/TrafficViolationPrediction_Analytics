```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("../data/raw/traffic_violations.csv")
df['observationDateTime'] = pd.to_datetime(df['observationDateTime'])

# Extract time features
df['hour'] = df['observationDateTime'].dt.hour
df['day_of_week'] = df['observationDateTime'].dt.dayofweek
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

# 1. Violations by Hour
plt.figure(figsize=(10, 5))
df['hour'].value_counts().sort_index().plot(kind='bar')
plt.title('Violations by Hour')
plt.xlabel('Hour')
plt.ylabel('Count')
plt.savefig('../data/processed/violations_by_hour.png')
plt.show()

# 2. Violations by Junction
plt.figure(figsize=(10, 5))
df['junctionName'].value_counts().plot(kind='barh')
plt.title('Violations by Junction')
plt.savefig('../data/processed/violations_by_junction.png')
plt.show()

# 3. Vehicle Type Distribution
plt.figure(figsize=(8, 5))
df['vehicleType'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Vehicle Type Distribution')
plt.savefig('../data/processed/vehicle_distribution.png')
plt.show()

print("EDA Complete!")
```
